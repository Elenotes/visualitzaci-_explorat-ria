> Treballem el **groupby()**

> La companyia amb una mitjana de retrassos més alta és la **"YV"**, amb 55 minuts.
    La que acomula menys retrassos és la **"AQ"** amb 21 minuts. Hi ha 10 companyies que acomules una mitjana de més de 3/4 d'hora de retràs.
    
> La companyia amb més vols **"WN"**, és la tercera que acomula menys retrassos.

```python
#Arrivals delay means by UniqueCarrier
grouped = new_df.groupby('UniqueCarrier')
arr_means = grouped['ArrDelay'].agg(np.mean).sort_values(ascending = False)
arr_means
```

> **agg()** method. Podem veure que les dades màximes i mínimes tenen valors estranys, amb la qual cosa, s'haurien de netejar abans de fer-ne les estadístiques.

```python
plt.figure(figsize=(7, 10))
plt.grid()
plt.xlabel('Delay mean (minutes)')
plt.ylabel('Unique Carrier')
plt.barh(arr_means.index, arr_means, color = 'green')
```

```python
new_df.agg(
    {"ArrDelay": ["min", "max", "median"],
     "DepDelay": ["min", "max", "median"],
     "velocitat": ["min", "max", "median"],
     "distancia_km":["min", "max", "median"]
    }
)
```
